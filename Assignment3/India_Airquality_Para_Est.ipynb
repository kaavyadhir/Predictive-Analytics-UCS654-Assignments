```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt

df = pd.read_csv('../input/india-air-quality-data/data.csv',encoding='latin1',low_memory=False)
x = pd.to_numeric(df['no2'], errors='coerce').dropna().values

r = 102317248

ar = 0.05 * (r % 7)
br = 0.3 * (r % 5 + 1)

z = x + ar * np.sin(br * x)

mu = np.mean(z)
sigma2 = np.var(z)

lam = 1 / (2 * sigma2)
c = math.sqrt(lam / math.pi)

print("Mean: ",mu)
print("Lambda: ",lam)
print("C: ",c)

z_sorted = np.sort(z)
pdf = c * np.exp(-lam * (z_sorted - mu)**2)

plt.plot(z_sorted, pdf)
plt.xlabel("z")
plt.ylabel("p(z)")
plt.show()
```
